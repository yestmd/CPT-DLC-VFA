## stage 3 training

```python
disk_path="put the disk directory here"
path=disk_path+"CPT_TOT_manuscript\\data processing\\fig\\stage 3 training"
training_latency=glob.glob(path+"\\*Latency*.csv")
```

```python
training_latency
```

```python
df=pd.DataFrame()
for i in training_latency:
    df_=pd.read_csv(i)
    df_group=df_.groupby(by=['gander', 'session_order', 'param']).agg(mean=pd.NamedAgg(column="value", aggfunc=np.mean), 
                                                                     std=pd.NamedAgg(column='value', aggfunc=np.std),
                                                                     count=pd.NamedAgg(column='value', aggfunc="count"))
    df_group['SEM']=df_group['std']/np.sqrt(df_group['count'])
    df_group=df_group.reset_index()
    df=pd.concat([df, df_group],ignore_index=True)
df=df.round(2)
```

```python
df["mean ± SEM"]=df['mean'].astype('str').str.pad(width=4,side='right',fillchar='0')+" ± "+df['SEM'].astype('str').str.pad(width=4,side='right',fillchar='0')
```

```python
df_paper=df[df['param'].isin(['Correct_Choice_Latency', 'Incorrect_Choice_Latency', 'Reward_Retrieval_Latency'])][["gander", "session_order", "param", "mean ± SEM"]].copy()
```

```python
df_supl=df[df['param'].isin(['Correct_Choice_Latency_sd', 'Incorrect_Choice_Latency_sd', 'Reward_Retrieval_Latency_sd'])][["gander", "session_order", "param", "mean ± SEM"]].copy()
```

```python
df_paper_pivot=df_paper.pivot(index=['session_order'], columns=['param','gander'], values=['mean ± SEM'])
df_paper_pivot.sort_index(axis=1, ascending=[False, True, False], inplace=True)
```

```python
df_paper_pivot.to_csv()
```

```python
df_supl_pivot=df_supl.pivot(index=['session_order'], columns=['param','gander'], values=['mean ± SEM'])
df_supl_pivot.sort_index(axis=1, ascending=[False, True, False], inplace=True)
```

```python
df_supl_pivot
```

```python
df=pd.DataFrame()
for i in TOT_latency:
    df_=pd.read_csv(i)
    df_group=df_.groupby(by=['gander', 'param']).agg(Mean=pd.NamedAgg(column="value", aggfunc=np.mean), 
                                                                     std=pd.NamedAgg(column='value', aggfunc=np.std),
                                                                     count=pd.NamedAgg(column='value', aggfunc="count"))
    df_group['SEM']=df_group['std']/np.sqrt(df_group['count'])
    df_group=df_group.reset_index()
    df=pd.concat([df, df_group],ignore_index=True)
df=df.round(2)
```

```python
df["mean ± SEM"]=df['Mean'].astype('str').str.pad(width=4,side='right',fillchar='0')+" ± "+df['SEM'].astype('str').str.pad(width=4,side='right',fillchar='0')
```

```python
bin_cond=[(df["param"].str.contains("bin_1")),(df["param"].str.contains("bin_2")),(df["param"].str.contains("bin_3")), 
         (df["param"].str.contains("bin_4")), (df["param"].str.contains("bin_5")),(df["param"].str.contains("bin_6"))]
bin_list=['bin 1','bin 2', 'bin 3', 'bin 4', 'bin 5', 'bin 6'] 

df['bin']=np.select(bin_cond, bin_list)
```

```python
sd_cond=[(df["param"].str.contains("sd")),(~df["param"].str.contains("sd"))]
sd_list=[' SD',''] 

df['sd']=np.select(sd_cond, sd_list)
```

```python
df['type']=df['latency']+ df['sd']
```
